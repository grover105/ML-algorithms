```python
import pandas as pd
df=pd.read_csv('car.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.mileage , df.selling_price)
```

```python
plt.scatter(df.Age , df.selling_price)
```

```python
dummies=pd.get_dummies(df.car_model)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['car_model','Mercedez Benz C class'],axis='columns')
final
```

```python
X=final.drop(['selling_price'],axis='columns')
X
```

```python
Y=final.selling_price
Y
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X,Y)
```

```python
model.score(X,Y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```
